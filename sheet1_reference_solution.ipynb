```python
import numpy as np 
from scipy.linalg import lu, lu_solve, lu_factor, cho_factor, cho_solve
```

```python
a=np.array([1,1,0])
b=np.array([-1,2,5])
M=np.array([[2,-1,0],[-1,2,-1],[0,-1,2]])
```

## a) Inner Product

```python
print(np.inner(a,b))
```

## b) Matrix-Vector Product

```python
print(M @ b)
```

## c) $l_2$ Norm

```python
print(np.linalg.norm(b,ord=None))
```

## d) Solve via LU

```python
lu_fac = lu_factor(M)
x_lu = lu_solve(lu_fac, b)
print(x_lu)
```

## e) Solve via Cholesky

```python
L_chol = np.linalg.cholesky(M)

# forward substitution
y_chol = np.linalg.solve(L_chol, b)

# bachward substitution
x_chol = np.linalg.solve(L_chol.T, y_chol)

print("x =", x_chol)
    
```

```python
print(np.abs(x_lu - x_chol))
```
