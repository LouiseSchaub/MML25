```python
# Programming Exercise 1
# Visualization using Matplotlib and NumPy

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# *** Part (a) ***
# Generate a random set of 2D points
np.random.seed(42)  # for reproducibility
points = np.random.rand(100, 2)  # 100 random 2D points

# Scatter plot of the random points
plt.figure()
plt.scatter(points[:, 0], points[:, 1], color='blue', alpha=0.6)
plt.title("Random 2D Points Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

# *** Part (b) ***
# Parameters of the 2D Gaussian distribution
mu = np.array([0, 1])
Sigma = np.array([[1, 0.25],
                  [0.25, 1]])

# Create a grid of (x, y) coordinates
x = np.linspace(-4, 4, 100)
y = np.linspace(-2, 6, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Compute the 2D Gaussian density
rv = multivariate_normal(mu, Sigma)
Z = rv.pdf(pos)

# Plot the Gaussian density as a contour plot
plt.figure()
plt.contourf(X, Y, Z, cmap='viridis')
#plt.xlim([-4,4])
#plt.ylim([-4,4])
plt.title(r"Density of 2D Gaussian Distribution $\mathcal{N}(\mu, \Sigma)$")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="Density")
plt.show()
```

```python
import plotly.graph_objects as go
import plotly.io as pio
from scipy.stats import multivariate_normal

for candidate in ["notebook_connected", "jupyterlab", "vscode", "colab", "iframe_connected", "browser"]:
    try:
        pio.renderers.default = candidate
        break
    except Exception:
        continue
#print("Plotly renderer set to:", pio.renderers.default)

# *** Part (a): random 2D points ***
np.random.seed(42)
points = np.random.rand(200, 2)  # a few more points for fun

scatter_fig = go.Figure(data=go.Scatter(x=points[:, 0], y=points[:, 1], mode="markers", marker=dict(size=6, opacity=0.7)))
scatter_fig.update_layout(title="Random 2D Points  Plotly", xaxis_title="X", yaxis_title="Y", template="plotly_white")

scatter_fig.show()
#scatter_fig.write_html("plotly_scatter_random_points.html", include_plotlyjs="cdn", auto_open=False)

# *** Part (b): 2D Gaussian density ***
mu = np.array([0, 1])
Sigma = np.array([[1.0, 0.25],
                  [0.25, 1.0]])

# grid
x = np.linspace(-4, 4, 200)
y = np.linspace(-2, 6, 200)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

rv = multivariate_normal(mu, Sigma)
Z = rv.pdf(pos)

contour_fig = go.Figure(data=go.Contour(x=x,y=y,z=Z, contours=dict(showlabels=True))) # This will give you interactive labels
contour_fig.update_layout(title=r"Density of 2D Gaussian $\mathcal{N}(\mu, \Sigma)$  Plotly", xaxis_title="X", yaxis_title="Y", template="plotly_white")

contour_fig.show()
#contour_fig.write_html("plotly_gaussian_density.html", include_plotlyjs="cdn", auto_open=False)

```
